# Check for Duplicate Rows

```python
## Check for duplicate rows
duplicate_rows = data_Migrants.duplicated()

# Display the number of duplicate rows
print(f"Number of duplicate rows: {duplicate_rows.sum()}")
```

```python
duplicates = data_Migrants[duplicate_rows]
print(duplicates.head())
```

```python
# Step 2: Remove duplicate rows (if any)
data_cleaned = data_Migrants.drop_duplicates()

# Display the shape of the data before and after removing duplicates
print(f"Original data shape: {data_Migrants.shape}")
print(f"Data shape after removing duplicates: {data_cleaned.shape}")
```

```python
## Check for duplicate rows
duplicate_rows = data_cleaned.duplicated()

# Display the number of duplicate rows
print(f"Number of duplicate rows: {duplicate_rows.sum()}")
```

```python
# Check for missing values in each column
missing_values = data_cleaned.isnull().sum()

# Display columns with missing values
print(missing_values[missing_values > 0])
```

#  Convert "Number of Dead" from float64 to int64 

```python
#  Convert "Number of Dead" from float64 to int64 using .loc[]
data_cleaned.loc[:, 'Number of Dead'] = data_cleaned['Number of Dead'].astype('int64')

# Verify the change
print(data_cleaned['Number of Dead'].dtype)
print(data_cleaned.head())
```

# Spliting "Coordinates" field   into latitude and longitude 

```python
# Step 1: Replace "Unknown" or other placeholders with NaN to handle them appropriately
data_cleaned.loc[:, 'Coordinates'] = data_cleaned['Coordinates'].replace('Unknown', pd.NA)

# Step 2: Split the "Coordinates" column into "Latitude" and "Longitude" using .loc[]
data_cleaned[['Latitude', 'Longitude']] = data_cleaned['Coordinates'].str.split(',', expand=True)

# Step 3: Convert the new columns to numeric data types using .loc[]
data_cleaned.loc[:, 'Latitude'] = pd.to_numeric(data_cleaned['Latitude'], errors='coerce')
data_cleaned.loc[:, 'Longitude'] = pd.to_numeric(data_cleaned['Longitude'], errors='coerce')

# Step 4: Optionally, drop the original "Coordinates" column if it's no longer needed
data_cleaned = data_cleaned.drop(columns=['Coordinates'])

# Step 5: Display the first few rows to verify the changes
print(data_cleaned.head())

# Step 6: Check for any remaining missing values in "Latitude" and "Longitude"
missing_lat_long = data_cleaned[['Latitude', 'Longitude']].isnull().sum()
print(missing_lat_long)
```

```python
# Find rows where either Latitude or Longitude is missing
missing_coords_rows = data_cleaned[data_cleaned['Latitude'].isnull() | data_cleaned['Longitude'].isnull()]
print(missing_coords_rows[[ 'Latitude', 'Longitude']])
```

```python
# Fill missing Latitude and Longitude values with 'Unknown'
data_cleaned['Latitude'] = data_cleaned['Latitude'].fillna('Unknown')
data_cleaned['Longitude'] = data_cleaned['Longitude'].fillna('Unknown')
```

```python
# Find rows where either Latitude or Longitude is missing
missing_coords_rows = data_cleaned[data_cleaned['Latitude'].isnull() | data_cleaned['Longitude'].isnull()]
print(missing_coords_rows[[ 'Latitude', 'Longitude']])
```

# Removing rows where "Minimum Estimated Number of Missing" and "Number of Survivors" has negative values

```python
#  Remove rows where "Minimum Estimated Number of Missing" has negative values
data_cleaned = data_cleaned[data_cleaned['Minimum Estimated Number of Missing'] >= 0]

# Remove rows where "Number of Survivors" has negative values
data_cleaned = data_cleaned[data_cleaned['Number of Survivors'] >= 0]
```
